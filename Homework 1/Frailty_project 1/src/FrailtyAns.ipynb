```python
import pandas  as pd
import seaborn    as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/clean_data.csv",encoding='ISO-8859-1')
print(data)
```

```python
#
data_N = data[data['Frailty']=='N']
data_Y = data[data['Frailty']=='Y']
print(data_N,data_Y,sep="\n")
```

```python
#we are Printing Max values in the dataset for all the columns where Frailty is Y,N
data_max = data.groupby(['Frailty'], sort=False).max()
data_max
```

```python
#we are Printing the max and min values in the dataset for the age Y,N
data_max_age = data.groupby(['Frailty'], sort=False)['Age'].max().reset_index()
data_min_age = data.groupby(['Frailty'], sort=False)['Age'].min().reset_index()

print(data_max_age,data_min_age,sep="\n")
```

```python
#we did the Barplot for Frailty count
plt.figure(figsize=(8,6))
sns.countplot(x="Frailty", data=data)
plt.show()
```

```python
# we are Plotting the barplot for Age and Height
from numpy import double
data['Age'] = data['Age'].astype(int)
data['Height'] = data['Height'].astype(double)
sns.barplot(data=data, x="Age",y="Height")
```

```python
plt.figure(figsize=(6,4))
sns.set_palette(sns.color_palette(('lightgreen','lightpink')))
sns.barplot(data=data, x="Age", hue='Frailty', y='Height');
sns.despine(top=True, right=True, left=False, bottom=False);
```
