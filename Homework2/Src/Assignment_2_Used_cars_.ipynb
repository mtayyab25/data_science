#Check for the Misising Values

#Mean of Kilometers driven

```python
mean_value = data['Kilometers_Driven'].mean()
print(f"Mean of Kilometers Driven: {mean_value}")
```

#Median of the seats

```python
median_seats =  data['Seats'].median()
print(f"Number of the seats: {median_seats}")
```

#Remove units from the columns and keep only the numerical values

```python
data['Mileage'] = data['Mileage'].astype(str)
data['Engine'] = data['Engine'].astype(str)
data['Power'] = data['Power'].astype(str)
data['New_Price'] = data['New_Price'].astype(str)

data['Mileage'] = data['Mileage'].str.replace(' km/kg', '').str.replace(' kmpl', '')
data['Mileage'] = pd.to_numeric(data['Mileage'], errors='coerce')

data['Engine'] = data['Engine'].str.replace(' CC', '')
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')

data['Power'] = data['Power'].str.replace(' bhp', '').str.replace('null', '0.0')
data['Power'] = pd.to_numeric(data['Power'], errors='coerce')

data['New_Price'] = data['New_Price'].str.replace(' Lakh', '')
data['New_Price'] = pd.to_numeric(data['New_Price'], errors='coerce')
```

```python
data.head()
```

```python
for column in missing_values.index:
    if missing_values[column] > 0:
        if data[column].dtype == np.number:
            # Replace missing values with the mean for numerical columns
            data[column].fillna(data[column].mean(), inplace=True)
        else:
            # Replace missing values with the mode for categorical columns
            data[column].fillna(data[column].mode().iloc[0], inplace=True)
```

Removing rows with missing values can result in a significant loss of data, so imputation ensures that the dataset remains complete and all columns have values. This approach maintains the statistical properties of the data, such as central tendency, making it suitable for various analysis and modeling tasks

```python
data['Kilometers_Driven'].fillna(data['Kilometers_Driven'].mean(), inplace=True)
data['Seats'].fillna(data['Seats'].median(), inplace=True)

```

```python
data['Fuel_Type'].fillna('Unknown', inplace=True)
data['Transmission'].fillna('Unknown', inplace=True)
```
